#### Dopasowanie zachłanne (*ang. greedy matching*)

Dopasowanie zachłanne dotyczy powtórzeń (czyli metaznaków `*`, `+`, `?`, '{}').

Wyrażenie `wn[0-9]*` oznacza ciąg 'wn', po którym bezpośrednio występuje 0 lub więcej cyfr. Do wzorca `wn[0-9]*` pasuje więc nie tylko fragment tekstu 'wn123', ale również 'wn', 'wn1' oraz 'wn12'. Pomimo to dla ciągu 'Bob Brown: bobbro`wn123`@home.co.uk' otrzymaliśmy wyłącznie jedno dopasowanie - 'wn123'. Takie zachowanie nazywamy **dopasowaniem zachłannym**. Dopasowanie zachłanne oznacza, że w przypadku niejednoznaczności dopasowania, dopasowywana jest zawsze maksymalna możliwa liczba powtórzeń. Jest to zachowanie domyślne dla wszystkich metaznaków opisujących liczbę powtórzeń (czyli `*`, `+`, `?`, `{}`). Możne je zmienić za pomocą odpowiedniej flagi.

#### Tryb wielolinijkowy (*arg. multiline*)

Tryb wielolnijkowy dotyczy metaznaków `^`, `$`.

Domyślny tryb to tryb wielolinijkowy. W trybie wielolinikowym metaznak `^` oznacza początek nowej linii, a metaznak `$` oznacza koniec nowej linii. To zachowanie można zmienić za pomocą odpowiedniej flagi. Po zmianie zachowania metaznak `^` oznaczać będzie początek całego tekstu, a metaznak `$` oznaczać będzie koniec całego tekstu.

#### 16. `()` - grupowanie wyrażeń 

Do tej pory wszystkie wyrażenia dotyczące powtórzeń dotyczyły pojedynczego znaku. Na przykład w wyrażeniu `wn[0-9]+` 'znak `+` oznacza jedno lub więcej powtórzeń znaku z zakresu 0-9 (czyli dowolnej cyfry). A co jeżeli chcielibyśmy napisać wzorzec, który szuka jednego lub więcej powtórzeń całego ciągu znaków `wn[0-9]`, a nie samej cyfry? Trzeba zgrupować ciąg, którego jedno lub więcej powtórzeń szukamy i znak `+` przykładamy do całej grupy. Grupujemy za pomocą nawiasów `()`. 

**Wyrażenie regularne**: `(wn[0-9])+`

#### Grupy przechwytujące (*ang. capturing groups*)

Zwróć uwagę, że poza samym dopasowaniem, oddzielnie dostajemy też dopasowanie samej grupy. Domyślnie grupy w wyrażeniach regularnych są tzw. grupami przechwytującymi. Oznacza to, że do dopasowanej grupy można się później odwołać. Takie odwołanie pozwala na jej ewentualne usunięcie lub zmodyfikowanie w oryginalnym tekście.

##### Przykład z większą liczbą grup w wyrażeniu

Wyrażenie regularne: `([a-zA-Z]+):\/\/([a-zA-Z]+)\.([a-zA-Z0-9]+)\.([a-zA-Z]{2,})`
