## JSON

Czym składniowo różni się json od słownika pythonowego ?
- w json akceptowalne są wyłącznie cudzysłowy (`"`), nie może być apostrofów (`'`)
- w json wartości boolean to `false` i `true`, a nie jak w pythonie `False`, `True`
- w json brak wartości to `null`, a nie jak w pythonie `None`

Typy obsługiwane w formacie json to:
- napis (w cudzysłowiu)
- liczba (może być int lub float)
- boolean (false, true)
- null
- listy (`[]`)
- object (`{}`)
