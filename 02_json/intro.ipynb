## JSON

Czym skadniowo rni si json od sownika pythonowego ?
- w json akceptowalne s wycznie cudzysowy (`"`), nie moe by apostrofw (`'`)
- w json wartoci boolean to `false` i `true`, a nie jak w pythonie `False`, `True`
- w json brak wartoci to `null`, a nie jak w pythonie `None`

Typy obsugiwane w formacie json to:
- napis (w cudzysowiu)
- liczba (moe by int lub float)
- boolean (false, true)
- null
- listy (`[]`)
- object (`{}`)
